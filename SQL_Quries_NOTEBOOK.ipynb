#q1_revenue_by_category

```python
%sql
SELECT
  Product_Category,
  ROUND(SUM(Total_Sales_INR), 2) AS total_revenue,
  COUNT(DISTINCT Order_ID) AS num_orders
FROM workspace.amazon_project.amazon_sales_clean
GROUP BY Product_Category
ORDER BY total_revenue DESC;
```

#q2_revenue_by_state

```python
%sql
SELECT
  State,
  ROUND(SUM(Total_Sales_INR), 2) AS total_revenue,
  COUNT(DISTINCT Order_ID) AS num_orders
FROM workspace.amazon_project.amazon_sales_clean
GROUP BY State
ORDER BY total_revenue DESC;
```

#q3_payment_method_mix


```python
%sql
SELECT
  Payment_Method,
  ROUND(SUM(Total_Sales_INR), 2) AS total_revenue,
  COUNT(DISTINCT Order_ID) AS num_orders
FROM workspace.amazon_project.amazon_sales_clean
GROUP BY Payment_Method
ORDER BY total_revenue DESC;
```

# q4_rating_by_category

```python
%sql
SELECT
  Product_Category,
  ROUND(AVG(Review_Rating), 2) AS avg_rating,
  COUNT(*) AS num_reviews
FROM workspace.amazon_project.amazon_sales_clean
GROUP BY Product_Category
HAVING COUNT(*) >= 10
ORDER BY avg_rating DESC;
```

# q5_rating_by_delivery_status

```python
%sql
SELECT
  Delivery_Status,
  ROUND(AVG(Review_Rating), 2) AS avg_rating,
  COUNT(*) AS num_orders
FROM workspace.amazon_project.amazon_sales_clean
GROUP BY Delivery_Status
ORDER BY avg_rating DESC;
```
